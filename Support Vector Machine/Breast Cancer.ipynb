```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
print(cancer['feature_names'])
```

```python
print(cancer['target'])
cancer['target_names']
```

```python
df=pd.DataFrame(data=cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.corr()
```

```python
fig=plt.figure(figsize=(8,6))
ax=fig.add_axes([0,0,1,1])
sns.heatmap(df.corr(),cmap='Blues',linewidth=2,linecolor="black")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,cancer['target'],test_size=0.3,random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001]} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
g_pred = grid.predict(X_test)
```

```python
print(classification_report(y_test,g_pred))
print(confusion_matrix(y_test,g_pred))
```
