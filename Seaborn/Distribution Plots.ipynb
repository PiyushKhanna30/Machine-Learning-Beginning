## Distribution Plots
Visualizing distribution of a data set.

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading in built dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe()
```

```python
tips.info()
```

# 1.Distplot
* The distplot shows the distribution of a univariate set of observations.

```python
fig,axes=plt.subplots(2,2,figsize=(14,10))

sns.distplot(tips['total_bill'],ax=axes[0][0]) #The y-axis in a density plot is the probability density function for the kernel density estimation.
sns.distplot(tips['total_bill'],bins=50,ax=axes[0][1])
sns.distplot(tips['total_bill'],hist=False,color='red',ax=axes[1][0])
sns.distplot(tips['total_bill'],kde=False,ax=axes[1][1])

plt.show()
```

# 2.Joinplot
jointplot() allows you to basically match up two distplots for bivariate data. **kind** parameter to compare with: 
* scatter
* reg
* resid 
* kde 
* hex

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],color='red',kind='scatter')
plt.show()
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='reg')
plt.show()
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='kde')
plt.show()
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='hex')
plt.show()
```

# 3.Pairplot
* pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips,markers='x')
plt.show
```

```python
sns.pairplot(tips,hue='smoker')
plt.show()
```

```python
sns.pairplot(tips,hue='sex',palette='magma')
plt.show()
```

# 4.Rugplot
* Draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot.

```python
fig,axes=plt.subplots(3,2,figsize=(14,14))

sns.rugplot(tips['size'],ax=axes[0][0])
sns.rugplot(tips['total_bill'],height=0.5,ax=axes[0][1])
sns.rugplot(tips['tip'],height=0.20,ax=axes[1][0])
sns.rugplot(tips['size'],axis='y',ax=axes[1][1])
sns.rugplot(tips['total_bill'],axis='y',height=0.15,ax=axes[2][0])
sns.rugplot(tips['tip'],axis='y',ax=axes[2][1])


plt.show()
```

# 5.Kdeplot
* kernel density estimation

```python
fig,axes=plt.subplots(3,3,figsize=(18,18))

sns.kdeplot(data=tips['size'],ax=axes[0][0])
sns.kdeplot(tips['total_bill'],ax=axes[0][1])
sns.kdeplot(tips['tip'],ax=axes[0][2])
sns.kdeplot(tips['size'],data2=tips['total_bill'],ax=axes[1][0])
sns.kdeplot(tips['total_bill'],data2=tips['tip'],ax=axes[1][1])
sns.kdeplot(tips['tip'],data2=tips['size'],ax=axes[1][2])

sns.kdeplot(data=tips['size'],ax=axes[2][0])
sns.rugplot(tips['size'],height=0.20,ax=axes[2][0])

sns.kdeplot(tips['total_bill'],ax=axes[2][1])
sns.rugplot(tips['total_bill'],height=0.20,ax=axes[2][1])

sns.kdeplot(tips['tip'],ax=axes[2][2])
sns.rugplot(tips['tip'],height=0.20,ax=axes[2][2])

plt.show()
```
