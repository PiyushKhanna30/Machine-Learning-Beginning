# Matrix Plots
Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data.
* Heatmap
* Clustermap

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# 1.Heatmap

```python
tips.corr()
```

```python
fig,ax = plt.subplots(3,1,figsize=(7,18))
sns.heatmap(tips.corr(),ax=ax[0])
sns.heatmap(tips.corr(),ax=ax[1],cmap='rainbow')
sns.heatmap(tips.corr(),ax=ax[2],cmap='coolwarm',annot=True)
plt.show()
```

```python
pv_flights = flights.pivot_table(index='month',values='passengers',columns='year')
pv_flights.head(12)
```

```python
fig,ax = plt.subplots(4,1,figsize=(7,24))
sns.heatmap(pv_flights,ax=ax[0])
sns.heatmap(pv_flights,ax=ax[1],cmap='rainbow',linewidth=1,linecolor='black')
sns.heatmap(pv_flights,ax=ax[2],cmap='coolwarm',annot=True)
sns.heatmap(pv_flights,ax=ax[3],cmap='coolwarm',annot=True,fmt='g')
plt.show()
```

# 2.Clustermap

```python
sns.clustermap(pv_flights)
```

```python
sns.clustermap(pv_flights,cmap='coolwarm',standard_scale=1)
```
