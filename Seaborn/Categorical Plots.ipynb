# Categorical Plots
* Barplot
* Countplot
* Boxplot
* Violinplot
* Stripplot
* Swarmplot
* Catplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

```python
tips.info()
```

```python
print(tips['day'].unique())
print(tips['sex'].unique())
print(tips['smoker'].unique())
print(tips['time'].unique())
```

# 1.Barplot
* To get aggregate data off a categorical feature in your data. **barplot** is a general plot that allows you to aggregate the categorical data based off some function, by default the mean.

```python
fig,ax=plt.subplots(4,2,figsize=(16,20))

sns.barplot(x='sex',y='total_bill',data=tips,ax=ax[0][0])
sns.barplot(x='sex',y='tip',data=tips,ax=ax[0][1],palette="coolwarm")
sns.barplot(x='time',y='total_bill',data=tips,ax=ax[1][0],palette="magma")
sns.barplot(x='time',y='tip',data=tips,ax=ax[1][1],palette="rainbow")
sns.barplot(x='smoker',y='total_bill',data=tips,ax=ax[2][0])
sns.barplot(x='smoker',y='tip',data=tips,ax=ax[2][1],palette="coolwarm")
sns.barplot(x='day',y='total_bill',data=tips,ax=ax[3][0],palette="magma")
sns.barplot(x='day',y='tip',data=tips,ax=ax[3][1],palette="rainbow")
plt.show()
```

```python
fig,ax=plt.subplots(4,2,figsize=(16,20))

sns.barplot(x='sex',y='total_bill',data=tips,ax=ax[0][0],estimator=np.std)    #now estimator instead of mean as standard deviation
sns.barplot(x='sex',y='tip',data=tips,dodge=False,ax=ax[0][1],hue="smoker",palette="coolwarm")
sns.barplot(x='time',y='total_bill',data=tips,ax=ax[1][0],palette="magma",estimator=np.sum,hue='day')
sns.barplot(x='time',y='tip',data=tips,ax=ax[1][1],hue='sex',palette="rainbow")
sns.barplot(x='smoker',y='total_bill',data=tips,ax=ax[2][0])
sns.barplot(x='smoker',y='tip',data=tips,ax=ax[2][1],palette="coolwarm")
sns.barplot(x='day',y='total_bill',data=tips,ax=ax[3][0],estimator=np.sum,palette="magma")
sns.barplot(x='day',y='tip',data=tips,ax=ax[3][1],hue='sex',estimator=np.sum,palette="rainbow")
plt.show()
```

# 2.Countplot
* This is same as barplot except the estimator is explicitly counting the number of occurrences. Which is why we only pass the x value.

```python
fig,ax=plt.subplots(4,2,figsize=(16,20))

sns.countplot(x='sex',data=tips,ax=ax[0][0],hue="smoker",palette="coolwarm")  
sns.countplot(x='sex',data=tips,ax=ax[0][1],palette="coolwarm")
sns.countplot(x='time',data=tips,ax=ax[1][0],palette="magma",hue='day')
sns.countplot(x='time',data=tips,ax=ax[1][1],hue='sex',palette="rainbow")
sns.countplot(x='smoker',data=tips,ax=ax[2][0],hue='sex')
sns.countplot(x='smoker',data=tips,ax=ax[2][1],palette="coolwarm")
sns.countplot(x='day',data=tips,ax=ax[3][0],palette="magma")
sns.countplot(x='day',data=tips,ax=ax[3][1],hue='sex',palette="rainbow")
plt.show()
```

# 3.Boxplot
* boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers using a method that is a function of the inter-quartile range.

```python
fig,ax=plt.subplots(4,2,figsize=(16,20))

sns.boxplot(x='sex',y='total_bill',data=tips,ax=ax[0][0])
sns.boxplot(x='sex',y='tip',data=tips,ax=ax[0][1],palette="coolwarm")
sns.boxplot(x='time',y='total_bill',data=tips,ax=ax[1][0],palette="magma",hue='sex')
sns.boxplot(x='time',y='tip',data=tips,ax=ax[1][1],palette="rainbow")
sns.boxplot(x='smoker',y='total_bill',data=tips,ax=ax[2][0])
sns.boxplot(x='smoker',y='tip',data=tips,ax=ax[2][1],palette="coolwarm")
sns.boxplot(x='day',y='total_bill',data=tips,ax=ax[3][0],palette="magma")
sns.boxplot(x='day',y='tip',data=tips,ax=ax[3][1],palette="rainbow")
plt.show()
```

```python
sns.boxplot(data=tips,palette='magma',orient='h')
```

# 4.Violinplot
* A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
fig,ax=plt.subplots(4,2,figsize=(16,20))

sns.violinplot(x='sex',y='total_bill',data=tips,ax=ax[0][0])
sns.violinplot(x='sex',y='tip',data=tips,ax=ax[0][1],palette="coolwarm")
sns.violinplot(x='time',y='total_bill',data=tips,ax=ax[1][0],palette="magma",hue='sex')
sns.violinplot(x='time',y='tip',data=tips,ax=ax[1][1],palette="rainbow")
sns.violinplot(x='smoker',y='total_bill',data=tips,ax=ax[2][0])
sns.violinplot(x='smoker',y='tip',data=tips,ax=ax[2][1],palette="coolwarm")
sns.violinplot(x='day',y='total_bill',data=tips,ax=ax[3][0],palette="magma",hue='sex')
sns.violinplot(x='day',y='tip',data=tips,ax=ax[3][1],palette="rainbow",hue='sex',split=True)
plt.show()
```

# 5.Stripplot 
* The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
sns.stripplot(x='tip',data=tips,jitter=True)
```

```python
fig,ax=plt.subplots(4,2,figsize=(16,20))

sns.stripplot(x='sex',y='total_bill',data=tips,ax=ax[0][0])
sns.stripplot(x='sex',y='tip',data=tips,ax=ax[0][1],palette="coolwarm",jitter=False)
sns.stripplot(x='time',y='total_bill',data=tips,ax=ax[1][0],palette="magma",hue='sex')
sns.stripplot(x='time',y='tip',data=tips,ax=ax[1][1],palette="rainbow",jitter=True,hue='sex',dodge=True)
sns.stripplot(x='smoker',y='total_bill',data=tips,ax=ax[2][0])
sns.stripplot(x='smoker',y='tip',data=tips,ax=ax[2][1],palette="coolwarm")
sns.stripplot(x='day',y='total_bill',data=tips,ax=ax[3][0],palette="magma",hue='sex')
sns.stripplot(x='day',y='tip',data=tips,ax=ax[3][1],palette="rainbow")
plt.show()
```

# 6.Swarmplot
* The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
fig,ax=plt.subplots(4,2,figsize=(16,20))

sns.swarmplot(x='sex',y='total_bill',data=tips,ax=ax[0][0])
sns.swarmplot(x='sex',y='tip',data=tips,ax=ax[0][1],palette="coolwarm")
sns.swarmplot(x='time',y='total_bill',data=tips,ax=ax[1][0],palette="magma",hue='sex')
sns.swarmplot(x='time',y='tip',data=tips,ax=ax[1][1],palette="rainbow",hue='sex',dodge=True)
sns.swarmplot(x='smoker',y='total_bill',data=tips,ax=ax[2][0],size=7,hue='sex')
sns.swarmplot(x='smoker',y='tip',data=tips,ax=ax[2][1],palette="coolwarm")
sns.swarmplot(x='day',y='total_bill',data=tips,ax=ax[3][0],palette="magma",hue='sex')
sns.swarmplot(x='day',y='tip',data=tips,ax=ax[3][1],palette="rainbow",hue='sex',dodge=True)
plt.show()
```

# 7.Catplot
* Cactorplot is the most general form of a categorical plot. It can take in a **kind** parameter to adjust the plot type.

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='bar')
sns.catplot(x='sex',data=tips,palette="coolwarm",kind='count')
sns.catplot(x='time',y='total_bill',data=tips,palette="magma",kind='box')
sns.catplot(x='time',y='tip',data=tips,palette="rainbow",kind='violin')
sns.catplot(x='smoker',y='total_bill',data=tips,kind='strip')
sns.catplot(x='smoker',y='tip',data=tips,palette="coolwarm",kind='swarm')
```

# Combing categorical plots

```python
fig=plt.figure(figsize=(14,10))
sns.violinplot(x="day", y="tip",data=tips,palette='rainbow')
sns.swarmplot(x="day", y="tip", data=tips,size=5,palette='Set1')
```

```python
fig=plt.figure(figsize=(14,10))
sns.violinplot(x="day", y="tip", hue='smoker',data=tips,palette='rainbow')
sns.swarmplot(x="day", y="tip", data=tips,size=5,hue='sex',palette='magma',dodge=True)
```
