```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<h1>CREATING DATASET USING SKLEARN</h1>

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
X,y = make_blobs(n_samples=600, n_features=2,centers=4, cluster_std=1.8,random_state=101)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, ax = plt.subplots(1, 2, sharey=True,figsize=(10,6))#sharey : shares y axis
ax[0].set_title('K Means')
ax[0].scatter(X[:,0],X[:,1],c=kmeans.labels_,cmap='rainbow')
ax[1].set_title("Original")
ax[1].scatter(X[:,0],X[:,1],c=y,cmap='rainbow')
ax[0].scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='black')
```

```python
kmeans.cluster_centers_[:,1]
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(X[:,0],X[:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(X[:,0],X[:,1],c=y,cmap='rainbow')
ax1.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='black')
```
