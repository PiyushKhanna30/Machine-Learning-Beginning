```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=[i for i in range (1,11)]
x=np.array(x)
y=[i*i for i in range(1,11)]
y=np.array(y)
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,label='x',ls='--',lw=2,marker='o')
axes.legend()
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.1,0.5,0.4,0.3])
axes1.plot(x,y,label='x',marker='o')
axes2.plot(y,x,'r',label='x*x')
axes1.legend()
axes2.legend()
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Title')
plt.show()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(8,6))
plt.tight_layout()
axes[0][0].plot(x,x)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(x,x)

axes[0][0].set_xlabel('X')
axes[0][0].set_ylabel('Y')
axes[0][0].set_title('x=[1,2,3,4...]')

axes[0][1].set_xlabel('X')
axes[0][1].set_ylabel('Y')
axes[0][1].set_title('y=[1,4,9,16,...]')

axes[1][0].set_xlabel('X')
axes[1][0].set_ylabel('Y')
axes[1][0].set_title('x')

axes[1][1].set_xlabel('X')
axes[1][1].set_ylabel('Y')
axes[1][1].set_title('y')

fig.savefig("my_fig.png",dpi=100)
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x,lw=2,ls='-.',color='red',alpha=0.6,marker='o')
axes.plot(x,x+2,lw=2,ls='--',color='blue',alpha=0.7,marker='+',markersize=10)
axes.plot(x,x+5,lw=3,ls=':',color='green',marker='s',markersize=10,markerfacecolor='yellow',markeredgewidth=4,markeredgecolor='black')
```
