```python
#Pie Chart
import matplotlib.pyplot as plt
lang='python','java','PHP','Javascript','C#','C++'
popularity=[22.7,17.6,8.8,8,7.7,6.7]
colors=['red','green','yellow','blue','orange','magenta']
#plt.pie(popularity)
#explode to put one of popularity outof say circle formed by pie chart
explode=(0.2,0.1,0,0,0,0)
#pie(popularity for formink pie,to highlight acc to explode,labels for slice,color to slice,put %age on slice,shadow for better visualization,start angle for pie)
plt.pie(popularity,explode=explode,labels=lang,colors=colors,autopct='%1.1f%%',shadow=True,startangle=140)
```
