```python
import numpy as np
#creates array from 0 to 10 with 25 numbers
x=np.linspace(0,10,25)
y=x*x+2

print(x)
print(y)
print(np.array([x,y]).reshape(25,2))
```

```python
from matplotlib import pyplot as plt
#creating canvas. here attribute inside figure is (figsize=(16,5))
fig=plt.figure()
```

```python
fig1=plt.figure()
#.add_axes(left,right,width,height) from 0 to 1. controls for canvas 
axes=fig1.add_axes([0.5,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes2=fig1.add_axes([0.5,0.1,0.8,0.8])
axes2.plot(y,x,'b')
```

```python
#creating subplots in an canvas
fig2,axes2=plt.subplots(nrows=1,ncols=2)
for ax in axes2:
    ax.plot(x,y,'r')
```

```python
fig3=plt.figure()
#Creating picture or graph inside another graph
axes3=fig3.add_axes([0.6,0.5,0.8,0.8])
axes3.plot(x,y,'g')

axes4=fig3.add_axes([0.1,0.4,0.4,0.4])
axes4.plot(y,x,'r--')
```

```python
#fig4=plt.figure(figsize=(16,9),dots per inches)more dpi more loading but for printing keep it 300
fig4=plt.figure(figsize=(16,9),dpi=100)
fig4.add_subplot()
plt.plot(x,y,'r')
fig4.add_subplot()
plt.plot(y,x,'b--')
```

```python
fig5,axes5=plt.subplots(figsize=(16,9),dpi=100) 
axes5.plot(x,y,'r')
```

```python
fig6,ax=plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,y,'r:')
ax[1].plot(y,x,'b--')
```

```python
fig6,ax=plt.subplots()
ax.plot(x,y,'r:')
ax.plot(y,x,'b--')
#Now putting title,x label,y label
ax.set_title('Title Is This')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
```

```python
fig7,ax=plt.subplots(figsize=(16,9))
#using marker ,marker size and markfacecolor for points that are ploted for making line
#marker - s,o,+ etc
ax.plot(x,x**2,'r:',label="x**2",marker='s',markersize=10,markerfacecolor='y')
#alpha for opacity
#lineweight or lw for the weight of line
#linestyle for styling of line
ax.plot(x,x**3,'b',label="x**3",alpha=0.5,lw=2,linestyle='--')
#Creating our own linestyle
#MIDE THE COMMA
line1,=ax.plot(x,x*200,'g',label="x*200")
#line.set_dashes([dash of len,skip len,second dash len,skip len])
line1.set_dashes([1,5,10,2])
#Now putting title,x label,y label
ax.set_title('Title Is This')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
#putting legend. Note label used within the .plot
ax.legend(loc=2, prop={'size': 20})

```

```python
fig7,ax=plt.subplots(1,2,figsize=(10,9))
#plottint multiple attributes in same line using single .plot
ax[0].plot(x,x**2,x**3,lw=2)
ax[0].grid(True)

#limiting the graph to be visualized or displyed

ax[1].plot(x,x**2,x**3,lw=2)
ax[1].grid(True)
ax[1].set_xlim([2,8])
ax[1].set_ylim([0,50])
```
