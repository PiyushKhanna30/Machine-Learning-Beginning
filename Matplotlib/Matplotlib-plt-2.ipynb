```python
import matplotlib.pyplot as plt
n=np.array([0,1,2,3,4,5])
fig,ax=plt.subplots(1,4,figsize=(16,5),dpi=100)

ax[0].set_title('Scatter')
ax[0].scatter(x,x+0.25*np.random.randn(len(x)))

ax[1].set_title("Step")
ax[1].step(n,n**2,lw=2)

ax[2].set_title("Bar")
ax[2].bar(n,n**2,align='center',width=0.5,alpha=0.5)

ax[3].set_title("Fill Between")
ax[3].fill_between(n,n*2,n*3,color='green',alpha=0.5)
ax[3].grid()
```

```python
#Radar Chart
fig=plt.figure(figsize=(16,9))
ax=fig.add_axes([0.0,0.0,0.6,0.6],polar=True)
#creting array using numpy beg from 0 to 2*pi with 100 numbers
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t,color='blue',lw=2)
```

```python
#Histogram
n1=np.random.randn(10000)
fig,ax=plt.subplots(1,2,figsize=(12,5))

ax[0].set_title('Default Histogram')
ax[0].hist(n1)

ax[1].set_title('Cumulative Histogram')
ax[1].hist(n1,cumulative=True,bins=5)
```
